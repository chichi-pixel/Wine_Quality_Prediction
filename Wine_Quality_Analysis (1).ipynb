```python
def null_cell(df): 
    total_missing_values = df.isnull().sum() 
    missing_values_per = df.isnull().sum()/df.isnull().count() 
    null_values = pd.concat([total_missing_values, missing_values_per], axis=1, keys=['total_null', 'total_null_perc']) 
    null_values = null_values.sort_values('total_null', ascending=False) 
    return null_values[null_values['total_null'] > 0]
```

```python
# Importing the required libraries
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Load the data
X = df[['fixed acidity','volatile acidity','sulphates','citric acid']] # feature matrix
y = df['quality'] # target vector

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create an instance of LogisticRegression
logistic_regression = LogisticRegression(solver='liblinear', multi_class='auto')

# Fit the model to the training data
logistic_regression.fit(X_train, y_train)

# Make predictions on the test data
y_pred = logistic_regression.predict(X_test)

# Evaluate the model performance
accuracy = logistic_regression.score(X_test, y_test)
print("Accuracy: ", accuracy)


# This code is created by ChatGPT! only feature matrix & target vector are added by human!:))
```
